```python
#Importar librerias 
#Visualisation libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
import datetime
import math
import matplotlib.mlab as mlab
import matplotlib
```

```python
#Importamos la DATA desde el PC
delincuencia = pd.read_excel("IDEFC_NM_ago22.xlsx")
delincuencia.head()
```

```python
#obtenemos la informacion de la DATA para saber que tipos de datos tenemos
delincuencia.info()
```

```python
#revisamos el numero de columnas y filas que tiene la data
delincuencia.shape
```

```python
#Eliminamos los NA del Data
delincuencia = delincuencia.fillna(0)
delincuencia
```

```python
#Revisamos que no tengamos ningun NA
delincuencia.isna().sum()
```

```python
#Convertimos los datos de Float a Int que para nuestro estudio los decimales no son validos
delincuencia[["Año","Clave_Ent","Enero","Febrero","Marzo", "Abril", "Mayo", "Junio", "Julio", "Agosto","Septiembre","Octubre","Noviembre","Diciembre", "Total de delitos"]] = delincuencia[["Año","Clave_Ent","Enero","Febrero","Marzo", "Abril", "Mayo", "Junio", "Julio", "Agosto","Septiembre","Octubre","Noviembre","Diciembre", "Total de delitos"]].astype(int) 
print(delincuencia.dtypes)
```

```python
#damos formato a la Columna de Año como date
delincuencia['Año'] = delincuencia['Año'].apply(pd.to_datetime,format='%Y')
```
